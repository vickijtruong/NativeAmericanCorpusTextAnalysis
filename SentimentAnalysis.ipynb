# Sentiment Analysis in Python

## Vader

```python
import os
os.chdir("/Users/vickijtruong/Documents/DH199/councils/")
all_files = os.listdir("/Users/vickijtruong/Documents/DH199/councils")
all_files.sort()
print(all_files)
```

```python
from nltk import tokenize
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer 

def sentiment(files):
    analyser = SentimentIntensityAnalyzer()
    snt = analyser.polarity_scores(files)
    print("{:-<40} {}".format(files, str(snt)))
```

```python
for file in all_files:
    with open(file,'r') as f:
        text = f.read()
        lines_list = tokenize.sent_tokenize(text)
        print(file)
        for line in lines_list:
            sentiment(line)
```

## TextBlob

```python
import os
os.chdir("/Users/vickijtruong/Documents/DH199/councils/")
all_files = os.listdir("/Users/vickijtruong/Documents/DH199/councils")
all_files.sort()
print(all_files)
```

```python
from nltk import tokenize
from textblob import TextBlob

for file in all_files:
    with open(file,'r') as f:
        text = f.read()
        lines_list = tokenize.sent_tokenize(text)
        print(file)
        blob = TextBlob(str(lines_list))
        for sentence in blob.sentences:
            print(sentence)
            print(sentence.sentiment.polarity)
            print(sentence.sentiment.subjectivity)


```
